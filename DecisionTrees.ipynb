```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier, plot_tree

# Synthetic user activity data (features: login frequency, session time)
X = np.array([[10, 5], [20, 7], [15, 6], [30, 10], [25, 9], [8, 4], [35, 11]])
y = np.array([0, 0, 0, 1, 1, 0, 1])  # 0: retained, 1: churned

# Train Decision Tree classifier
clf = DecisionTreeClassifier(max_depth=2, random_state=42)
clf.fit(X, y)

# Visualize the decision tree
plt.figure(figsize=(10, 6))
plot_tree(clf, feature_names=['Login Frequency', 'Session Time'], class_names=['Retained', 'Churned'], filled=True)
plt.title('Decision Tree for User Churn Prediction')
plt.show()
```
